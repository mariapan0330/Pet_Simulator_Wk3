### Exercise 2 - Write a Python class for an Animal that has a name and energy attributes. The animal class should also have methods for eat, sleep, and play that will take in an integer and increase/decrease the energy of the animal with a formatted print statement

```python
# Example 1
# buddy = Animal('Buddy', 10)
# buddy.play(5) -> "Buddy is playing for 5 minutes. His energy is now 5"
# buddy.sleep(10) -> "Buddy is sleeping for 5 minutes. His energy is now 15"
```

```python
from IPython.display import clear_output
import time
```

```python
class Animal():
    def __init__(self, name, energy=100):
        self.name = name
        self.energy = energy
    
    def inactive(self, first=False):
        clear_output()
        if first:
            print(f"Meet {self.name}.")
            
        # Thank you, Timur (github: timurmalkoc) for the inspiration for this:
        print("""
    *************************
    *                       *
    *                       *
    *                       *
    *      /\___/\          *
    *     / .   . \         *
    *     \___*___/         *
    *                       *
    *************************
                """)
        print(f"ENERGY: {self.energy}")
        
        while True:
            action = input("What would you like to do? EAT / SLEEP / PLAY: ").lower().strip()
            if action not in {'eat','sleep','play'}:
                print("That didn't work.")
            else:
                break
        
        while True:
            duration = input(f"How long would you like to {action}?")
            if duration.isdigit() == False:
                print("That didn't work.")
            else:
                break
                
        if action == 'eat':
            self.eat(int(duration))
        if action == 'sleep':
            self.sleep(int(duration))
        if action == 'play':
            self.play(int(duration))

    def eat(self, duration):
        clear_output()
        print("""
    *************************
    *                       *
    *                       *
    *                       *
    *      /\___/\          *
    *     /  .   .\         *
    *     \_____O_/ ***     *
    *                       *
    *************************
                """)
        self.energy += duration
        print(f"ENERGY: {self.energy}")
        
#         times_eaten_in_a_row = 0
#         if self.energy <= 50:
        print(f"{self.name.title()} has eaten for {duration} minutes." +
              f" Energy is now at {self.energy}.")
        any_key = input("ENTER ANY KEY to continue! ")
        self.inactive()
#         elif self.energy > 200:
            

    def sleep(self, duration):
        clear_output()
        print("""
    *************************
    *                 Z     *
    *                z      *
    *              z        *
    *      /\___/\          *
    *     / _   _ \         *
    *     \___*___/         *
    *                       *
    *************************
                """)
        self.energy += duration
        print(f"ENERGY: {self.energy}")
        print(f"{self.name.title()} has slept for {duration} minutes." +
              f" Energy is now at {self.energy}.")
        any_key = input("ENTER ANY KEY to continue! ")
        self.inactive()
    

    def play(self, duration):
        clear_output()
        print("""
    *************************
    * now playing my        *
    * favorite game:        *
    * Knock This Glass Over *
    *      /\___/\          *
    *     / O   O \         *
    *     \___*___/   /---\ *
    *                 \___/ *
    *************************
                """)
        self.energy -= duration
        print(f"ENERGY: {self.energy}")
        if self.energy > 10:
            print(f"{self.name.title()} has played for {duration} minutes." + 
                  f" Energy is now at {self.energy}.")
            any_key = input("ENTER ANY KEY to continue! ")
            self.inactive()
        elif self.energy <= 10 and self.energy > 0:
            print(f"Careful! {self.name.title()} has played for {duration} minutes"+
                  f" and now has only {self.energy} energy."+
                  " If it plays too long, it won't make it.")
            any_key = input("ENTER ANY KEY to continue! ")
            self.inactive()
        elif self.energy <= 0:
            clear_output()
            print("""
    *************************
    *                       *
    *    GAME               *
    *        OVER           *
    *                       *
    *      /\___/\          *
    *     / x   X \         *
    *     \___*___/         *
    *************************
                """)
            print(f"ENERGY: 0!!")
            print(f"Oh no! {self.name} has died. :,(")
            any_key = input("ENTER ANY KEY to quit!")
```

```python
def start():
    print("Welcome to the Mini Pet Simulator!")
    name = input("What would you like to name your new pet? ").strip().title()
    print("That's a very good name for one very good pet!")
    any_key = input(f"ENTER ANY KEY to continue and meet {name}! ") 
    # ^This is a very fun phrase when you name your pet "God".
    
    pet = Animal(name)
    pet.inactive(True)
```
